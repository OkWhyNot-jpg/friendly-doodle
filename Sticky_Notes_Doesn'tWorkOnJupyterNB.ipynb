```python
import ipywidgets as widgets
from IPython.display import display

class StickyNotesApp:
    def __init__(self):
        # Create a list to store the notes
        self.notes = []

        # Create the GUI elements
        self.note_text = widgets.Textarea(height=10, width=50)
        self.create_button = widgets.Button(description="Create Note")
        self.view_button = widgets.Button(description="View Notes")
        self.delete_button = widgets.Button(description="Delete Notes")

        # Register button click event handlers
        self.create_button.on_click(self.create_note)
        self.view_button.on_click(self.view_notes)
        self.delete_button.on_click(self.delete_notes)

        # Display the GUI elements
        display(self.note_text, self.create_button, self.view_button, self.delete_button)

    def create_note(self, _):
        # Get the text from the note_text widget
        note = self.note_text.value

        # Add the note to the list
        self.notes.append(note)

        # Clear the note_text widget
        self.note_text.value = ""

        print("Note created successfully!")

    def view_notes(self, _):
        # Create a text widget to display the notes
        notes_text = widgets.Textarea(height=10, width=50)

        # Insert each note into the text widget
        for note in self.notes:
            notes_text.value += note + "\n"

        # Display the notes
        display(notes_text)

    def delete_notes(self, _):
        # Clear the notes list
        self.notes = []

        print("Notes deleted successfully!")


# Create an instance of the StickyNotesApp
app = StickyNotesApp()
```
